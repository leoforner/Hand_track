```python
#@title Trabalho planilha rpg  combate V1{ vertical-output: true, display-mode: "form" }
class Personagem:
    def __init__(self, nome, classe):
        self.__nome = nome
        self.__classe = classe
        self.__vida = 100
        self.__mana = 50
    
    def __str__(self):
        return f"{self.__nome}, {self.__classe} ({self.__vida} HP, {self.__mana} MP)"
    
    def atacar(self):
        print(f"{self.__nome} atacou com um ataque básico!")
    
    def defender(self):
        print(f"{self.__nome} se defendeu!")
    
    def usar_habilidade(self, habilidade):
        print(f"{self.__nome} usou a habilidade {habilidade}!")
    
    def tomar_dano(self, dano):
        self.__vida -= dano
        if self.__vida <= 0:
            print(f"{self.__nome} foi derrotado!")
    
    def recuperar_vida(self, cura):
        self.__vida = min(self.__vida + cura, 100)
    
    def recuperar_mana(self, mana):
        self.__mana = min(self.__mana + mana, 50)
    
    def get_nome(self):
        return self.__nome
    
    def get_classe(self):
        return self.__classe
    
    def get_vida(self):
        return self.__vida
    
    def get_mana(self):
        return self.__mana


class Guerreiro(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Guerreiro")
    
    def atacar(self):
        print(f"{self.get_nome()} atacou com um golpe de espada!")
    
    def defender(self):
        print(f"{self.get_nome()} bloqueou o ataque com o escudo!")
    
    def usar_habilidade(self, habilidade):
        if habilidade == "Berserk":
            print(f"{self.get_nome()} usou a habilidade Berserk e ficou mais forte!")
            self.recuperar_vida(10)
        elif habilidade == "Grito de Guerra":
            print(f"{self.get_nome()} usou o Grito de Guerra e inspirou seus aliados!")
            self.recuperar_mana(10)
        else:
            super().usar_habilidade(habilidade)


class Mago(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Mago")
    
    def atacar(self):
        print(f"{self.get_nome()} lançou uma bola de fogo!")
        self.recuperar_mana(5)
    
    def defender(self):
        print(f"{self.get_nome()} conjurou um escudo mágico!")
        self.recuperar_mana(5)
    
    def usar_habilidade(self, habilidade):
        if habilidade == "Teleporte":
            print(f"{self.get_nome()} usou a habilidade Teleporte e desapareceu!")
        elif habilidade == "Cura":
            print(f"{self.get_nome()} usou a habilidade Cura e se curou!")
            self.recuperar_vida(10)
            self.recuperar_mana(10)
        else:
            super().usar_habilidade(habilidade)


# Testando as classes
guerreiro1 = Guerreiro("Conan")
mago1 = Mago("Gandalf")

print(guerreiro1)
print(mago1)

guerreiro1.atacar()
mago1.atacar()

guerreiro1.defender()
mago1.defender()

guerreiro1.usar_habilidade("Berserk")
mago1.usar_habilidade("Cura")

guerreiro1.tomar_dano(50)
mago1.tomar_dano(70)

print(guerreiro1)
print(mago1)

```

```python
#@title Trabalho planilha rpg V2{ vertical-output: true, display-mode: "form" }


class Personagem:
    def __init__(self, nome, raca, classe, nivel):
        self.nome = nome
        self.raca = raca
        self.classe = classe
        self.nivel = nivel
        self.pontos_de_vida = self.calcular_pv()
        self.modificador_forca = self.calcular_modificador(self.atributos["forca"])
        self.modificador_destreza = self.calcular_modificador(self.atributos["destreza"])
        self.modificador_constituicao = self.calcular_modificador(self.atributos["constituicao"])
        self.modificador_inteligencia = self.calcular_modificador(self.atributos["inteligencia"])
        self.modificador_sabedoria = self.calcular_modificador(self.atributos["sabedoria"])
        self.modificador_carisma = self.calcular_modificador(self.atributos["carisma"])
    
    # Atributos base de cada raça
    atributos_base = {
        "anao": {"forca": 2, "constituicao": 2, "destreza": -2},
        "elfo": {"destreza": 2, "inteligencia": 2, "constituicao": -2},
        "halfling": {"destreza": 2, "carisma": 2, "forca": -2},
        "humano": {"nenhum": 0},
        "meio-elfo": {"nenhum": 0},
        "meio-orc": {"forca": 2, "constituicao": 2, "carisma": -2},
        "gnomo": {"inteligencia": 2, "destreza": 2, "forca": -2},
        "draconato": {"forca": 2, "carisma": 1},
        "tiefling": {"carisma": 2, "inteligencia": 1, "constituicao": -1}
    }
    
    # Dados base de cada classe
    dados_base = {
        "barbaro": {"dado_vida": 12, "pontos_vida_nivel": 7},
        "bardo": {"dado_vida": 8, "pontos_vida_nivel": 5},
        "clerigo": {"dado_vida": 8, "pontos_vida_nivel": 5},
        "druida": {"dado_vida": 8, "pontos_vida_nivel": 5},
        "feiticeiro": {"dado_vida": 6, "pontos_vida_nivel": 4},
        "guerreiro": {"dado_vida": 10, "pontos_vida_nivel": 6},
        "ladino": {"dado_vida": 8, "pontos_vida_nivel": 5},
        "mago": {"dado_vida": 6, "pontos_vida_nivel": 4},
        "monge": {"dado_vida": 8, "pontos_vida_nivel": 5},
        "paladino": {"dado_vida": 10, "pontos_vida_nivel": 6},
        "patrulheiro": {"dado_vida": 10, "pontos_vida_nivel": 6}
    }
    
    def calcular_pv(self):
        dados = self.dados_base[self.classe]
        pontos_vida = dados["pontos_vida_nivel"] * self.nivel
        return pontos_vida + dados["dado_vida"]
    
    def calcular_modificador(self, valor_atributo):
        return (valor_atributo - 10) // 2
    
    def gerar_ficha(self):
        print("----- FICHA DE PERSONAGEM -----")
        print(f"Nome: {self.nome}")
        print(f"Raça: {self.raca}")
        print(f"Classe: {self.classe}")
        print(f"Nível: {self.nivel}")
        print(f"Pontos de Vida: {self.pontos_de_vida}")
        print(f"Modificador de Força: {self.modificador_forca}")
        print(f"Modificador de Destreza: {self.modificador_destreza}")
        print(f"Modificador de Constituição: {self.modificador_constituicao}")
        print(f"Modificador de Inteligência: {self.modificador_inteligencia}")
        print(f"Modificador de Sabedoria: {self.modificador_sabedoria}")
        print(f"Modificador de Carisma: {self.modificador_carisma}")
        print("------------------------------")
        
# Criação do personagem
nome = input("Digite o nome do personagem: ")
raca = input("Digite a raça do personagem: ")
classe = input("Digite a classe do personagem: ")
nivel = int(input("Digite o nível do personagem: "))

personagem = Personagem(nome, raca, classe, nivel)
personagem.gerar_ficha()
```
